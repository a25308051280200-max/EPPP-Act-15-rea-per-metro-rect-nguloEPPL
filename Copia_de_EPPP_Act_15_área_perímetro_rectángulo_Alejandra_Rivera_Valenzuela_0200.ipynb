📝 Redacción del Problema
* Desarrollar un programa en Python que permita al usuario realizar cálculos geométricos básicos sobre un rectángulo. El sistema debe mostrar las fórmulas matemáticas correspondientes, presentar un menú interactivo y permitir el cálculo del área y el perímetro mediante funciones específicas, utilizando estructuras de control como while, if, elif y else.

📐 Fórmulas a utilizarPara un rectángulo con base
* $b$ y altura $h$:Área:
* $A = b \times h$Perímetro:
* $P = 2 \times (b + h)$ o $P = 2b + 2h$

* Modularidad: Cada cálculo tiene su propia función (def), lo que hace el código limpio.

* Ciclo Infinito: El while True permite que el usuario haga múltiples cálculos sin tener que reiniciar el programa.

* Validación Simple: He añadido un bloque try-except por si el usuario escribe letras en lugar de números por accidente, evitando que el programa "truene".

```python
# ==================================
# Problema de cálculo geométrico: Rectángulo
# Elaborado por: [ALEJANDRA RIVERA VALENZUELA]
# Fecha: 2026
# =====================================

def mostrar_formulas():
    """Muestra las fórmulas matemáticas del rectángulo."""
    print("\n" + "="*40)
    print("📐 FÓRMULAS DEL RECTÁNGULO 📐")
    print("="*40)
    print("1. Área del Rectángulo:")
    print("   Fórmula: A = base * altura")
    print("\n2. Perímetro del Rectángulo:")
    print("   Fórmula: P = 2 * (base + altura)")
    print("="*40 + "\n")

def calcular_area_rectangulo():
    """Solicita datos y calcula el área."""
    print("\n🟦 CÁLCULO DEL ÁREA")
    base = float(input("Ingresa la base del rectángulo: "))
    altura = float(input("Ingresa la altura del rectángulo: "))
    area = base * altura
    print(f"✅ El área es: {area:.2f}")

def calcular_perimetro_rectangulo():
    """Solicita datos y calcula el perímetro."""
    print("\n📏 CÁLCULO DEL PERÍMETRO")
    base = float(input("Ingresa la base del rectángulo: "))
    altura = float(input("Ingresa la altura del rectángulo: "))
    perimetro = 2 * (base + altura)
    print(f"✅ El perímetro es: {perimetro:.2f}")

def menu():
    """Muestra el menú y retorna la opción."""
    print("\n" + "*"*30)
    print("🌟 MENÚ RECTÁNGULO 🌟")
    print("*"*30)
    print("1️⃣  Calcular Área")
    print("2️⃣  Calcular Perímetro")
    print("3️⃣  Salir")
    print("*"*30)
    opcion = int(input("👉 Elige una opción (1-3): "))
    return opcion

# --- Programa Principal ---

def main():
    # Mostramos fórmulas una vez al inicio
    mostrar_formulas()

    # Bucle principal
    while True:
        try:
            opcion_seleccionada = menu()

            if opcion_seleccionada == 1:
                calcular_area_rectangulo()
            elif opcion_seleccionada == 2:
                calcular_perimetro_rectangulo()
            elif opcion_seleccionada == 3:
                print("\n👋 ¡Programa finalizado! Éxito en tus estudios. 🚀\n")
                break  # Rompe el ciclo while para salir
            else:
                print("\n⚠️ Opción no válida. Intenta con 1, 2 o 3.")

        except ValueError:
            print("\n❌ Error: Por favor, ingresa solo números.")

# Ejecutar el programa
if __name__ == "__main__":
    main()
```
